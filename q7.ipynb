```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("us-marriages-divorces-1867-2014.csv", usecols=['Year','Marriages_per_1000','Divorces_per_1000'])
cf = df.loc[df['Year'].isin([1900,1950,2000])]
cf
```

```python
cf.plot.bar(x='Year')
plt.ylabel('Per 1000 people')
plt.title('Marriages and divorces per capita for 1900, 1950, and 2000')
plt.show()
```
