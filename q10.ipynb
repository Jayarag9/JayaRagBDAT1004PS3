```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")
df
```

```python
df.plot.scatter(x = 'Total Arcade Revenue (billions)', y = 'Computer Science Doctorates Awarded (US)', c = 'Year')
plt.title('Relationship between the total revenue \n earned by arcades and the number of Computer Science \n PhDs awarded in the U.S. between 2000 and 2009')
plt.show()
```
